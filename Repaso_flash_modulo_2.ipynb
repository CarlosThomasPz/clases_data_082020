# Pruebas de hipotesis

# Los estudiantes que tienen internet tiene una mejor calificacion promedio por arriba de %80 
# H0 = x > 80
# H1 = x <= 80 #esta es la que buscamos

Recuerda, el alpha esta dado por el nivel de confianza:
  + 99% , el alpha es de 0.01
  + 95% , el alpha es de 0.05
  + 90%, el alpha es de 0.1

```python
#H0 puede ser de una cola, para saber si es mayor o menor al valor esperado,H1 seria menor o mayor que lo observado
#H0 es igual al valor esperado, H1 es diferente del valor esperado
```

```python
students['internet'].unique()
```

```python
int_y = students[students['internet'] == 'yes'][['G1','G2','G3']]
```

```python
int_y['prom'] = (int_y['G1']+int_y['G2']+int_y['G3'])/3
```

```python
int_y['prom'] = int_y['prom']*5 #escalado a califacion de 100
int_y['prom']
```

```python
import statsmodels.api as sm
```

```python
help(sm.stats.ztest)
```

```python
sm.stats.ztest(int_y['prom'], value = 80, alternative = 'smaller')
```

```python
#p_value se contrasta con nuestro alpha que definimos
```

```python
url = "https://raw.githubusercontent.com/kshedden/statswpy/master/NHANES/merged/nhanes_2015_2016.csv"
da = pd.read_csv(url)
da.head()
```

```python
da.columns
```

```python
da['RIAGENDR'].unique()
```

```python
mujeres = da[da['RIAGENDR'] == 1]
hombres = da[da['RIAGENDR'] == 2]
```

```python
n1 = len(mujeres)
media_m= mujeres['BMXBMI'].mean()
std_m = mujeres['BMXBMI'].std()

print(n1,media_m, std_m)
```

```python
n2 = len(hombres)
media_h = hombres['BMXBMI'].mean()
std_h = hombres['BMXBMI'].std()

print(n2,media_h, std_h)
```

```python
#H0  = el indice de masa corporal es igual entre hombres y mujeres 
#H1  = el indice de masa corporal es diferente  entre hombre y mujeres
```

```python
results = sm.stats.ztest(mujeres['BMXBMI'].dropna(),hombres['BMXBMI'].dropna(), alternative = 'two-sided' )
#alternative es H0, podemos evaluar si es mayor, si es menor o si son diferentes
results
```

```python
#tenemos un p_value de 0.000000000006 como es menor a cualquier alpha sea 99%,95%,%90
#RECHAZAMOS H0 es decir los hombres y las mujeres tienen diferentes medias de indice de masa corporal
```
