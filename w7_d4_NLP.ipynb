# TSNE 

```python
from sklearn.manifold import TSNE
```

```python
from sklearn.datasets import load_iris
```

```python
import matplotlib.pyplot as plt
```

```python
x, y = load_iris(return_X_y = True)
```

```python
fig, ax = plt.subplots(1,4, figsize =(15,7))

for i,p in enumerate([30,40,50,100]):
    tsne  = TSNE(perplexity = p)
    
    x_emb = tsne.fit_transform(x)
    
    ax[i].scatter(x_emb[:,0], x_emb[:,1], c = y)
    ax[i].set_title(f'Perplexity = {p}')
    
plt.show();
```

```python
import nltk
```

```python
from nltk import sent_tokenize
from nltk import word_tokenize
```

```python
with open('data/Natural_Language_Processing_Text.txt','r') as f :
    text = f.read()
```

```python
text
```

```python
nltk.download('punkt')
```

```python
sentences = sent_tokenize(text)

sentences
```

```python
words = word_tokenize(text)
words
```

```python
from nltk.probability import FreqDist
```

```python
fdist = FreqDist(words)

fdist.most_common(10)
```

```python
fdist.plot(10)
```

```python
words_no_punc = []

for w in words:
    if w.isalpha():
        words_no_punc.append(w.lower())
    
    
words_no_punc
```

```python
fdist= FreqDist(words_no_punc)

fdist.plot(10)
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords

stopwords_en = stopwords.words('english') 
#son aquellas palabras que no tiene un peso en el contexto del texto, es decir que no apórtan informacion relevante

print(stopwords_en)
```

```python
stopwords_es = stopwords.words('spanish')
```

```python
clean_words = []

for w in words_no_punc:
    if w not in stopwords_en:
        clean_words.append(w)
        
clean_words
```

```python
fdist = FreqDist(clean_words)

fdist.most_common(10)
```

```python
fdist.plot(10)
```

```python
from wordcloud import WordCloud
```

```python
clean_words
```

```python
text_clean = " ".join(clean_words)
```

```python
word_cloud = WordCloud().generate(text_clean)

plt.figure(figsize = (12,12))

plt.imshow(word_cloud)

plt.axis('off')

plt.show();
```

```python
from nltk.stem import PorterStemmer

porter = PorterStemmer()

lista_palabras = ['study', 'studied','studying', 'studies']
```

```python
for p in lista_palabras:
    print(porter.stem(p))
```

```python
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer

lem =  WordNetLemmatizer()
```

```python
for p in lista_palabras:
    print(lem.lemmatize(p, pos = 'v'))
```

```python
lem.lemmatize('studies')
```

```python
lem.lemmatize('studied')
```

```python
lista_2 = ['am','is','are', 'were']

for e in lista_2:
    print(lem.lemmatize(e, pos = 'v'))
```

```python
lem.lemmatize('sillas', pos = 'n')
```

```python
from nltk.corpus import names
```

```python
nltk.download('names')
```

```python
names = ([(name, 'male') for name in names.words('male.txt')] + [(name, 'female') for name in names.words('female.txt')])
```

```python
len(names)
```

```python
def gender_features(word):
    return {'last_letter': word[-2:]} 
```

```python
featuresets = [(gender_features(n),g ) for (n,g) in names]
```

```python
featuresets
```

```python
from nltk.classify import NaiveBayesClassifier
import nltk.classify.util
```

```python
train = featuresets
classifer = nltk.NaiveBayesClassifier.train(train)
```

```python
print(classifer.classify(gender_features('Frank')))
```

```python
name  = input('Nombre: ')
print(classifer.classify(gender_features(name)))
```
