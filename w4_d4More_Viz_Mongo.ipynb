```python
import os
lst_cryp = ['data/crypto/'+file for file in os.listdir("data/crypto/") if file.endswith(".csv")]
```

```python
lst_cryp
```

```python
cryp_df = [pd.read_csv(e) for e in lst_cryp]
```

```python
cryp_df = [e[['Date','High']] for e in cryp_df]
```

```python
df_final = reduce(lambda left,right: pd.merge(left,right,on='Date'), cryp_df)
```

```python
df_final
```

```python
df_final
```

```python
df_final.sort_index(ascending=True,inplace=True)
```

```python
lst_names = [re.search('(?<=-)\w+(-\w+)?',e).group() for e in lst_cryp]
```

```python
lst_names
```

```python
re.search('(?<=foo)\w+','mksfoobarasdwe').group()
```

```python
date = lambda x: datetime.strptime(x, '%b %d, %Y')
```

```python
df_final['Date'] = df_final['Date'].apply(date)
df_final
```

```python
df_final.set_index('Date', inplace = True)
```

```python
df_final.columns = lst_names
```

```python
df_final
```

```python
clean = lambda x: x.replace(',','')
```

```python
for i in df_final.columns:
    if df_final[i].dtype == np.object:
        df_final[i] = df_final[i].apply(clean)
```

```python
for i in df_final.columns:
    if df_final[i].dtype == np.object:
        df_final[i] = df_final[i].astype('float64')
```

```python
df_final.to_csv('data/crypto_complete.csv')
```
