#  Group by - summarization

```python
carros = pd.read_csv('data/vehicles.csv')
```

```python
carros
```

```python
carros.info()
```

```python
carros['Transmission'].unique()
```

```python
carros.isna().sum()
```

```python
import re
```

```python
re.findall
```

```python
spd = lambda x: re.findall(r'\d',x)[0] if bool(re.search(r'\d',x)) else '0'
```

```python
carros['Transmission']= carros['Transmission'].apply(spd)
```

```python
carros.groupby('Transmission').sum()
```

```python
carros.groupby('Transmission').mean()
```

```python
carros.groupby('Transmission')['Year','Cylinders','Combined MPG'].mean()
```

```python
carros.groupby('Transmission')['Year','Cylinders','Combined MPG'].agg(['mean','std','sum'])
```

```python
carro_g = carros.groupby(['Transmission','Cylinders'])['Combined MPG','Highway MPG'].mean()
carro_g
```

```python
carro_g.loc['3',4]
```

```python
carro_g.iloc[2]
```

```python
carro_g.iloc[[7]]
```

```python
carro_g.iloc[[6],[1]]
```

```python
carro_g = carros.groupby(['Transmission','Cylinders'])['Combined MPG','Highway MPG'].agg(['mean','std'])
carro_g
```

```python
from scipy import stats

def agg_mode(x):
    return(stats.mode(x)[0])
```

```python
carros.groupby('Year')['Vehicle Class'].agg(agg_mode)
```

# Correlacion 
La correlacion mide el grado en que dos variables estan relaciones mediante la fuerza y la dirrecion de una relacion lineal y proporcionalidad

```python
carros['City MPG'].corr(carros['Highway MPG'])
```

```python
carros['City MPG'].corr(carros['City MPG'])
```

```python
carros.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set(style="white")
corr = carros.corr(method = 'spearman')
mask = np.triu(np.ones_like(corr, dtype=np.bool))
```

```python
mask
```

```python
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(150, 100, as_cmap=True)

sns.heatmap(corr, cmap=cmap, annot=True, mask = mask)
```
