```python
def square(L, c=(0, 0)):
    x=(np.array([0,0,1,1, 0])*L)-c[0]-L/2
    y=(np.array([0,1,1,0, 0])*L)-c[1]-L/2
    print(x,y)
    return np.array([x, y])
```

```python
c = square(10, (0,0))
c
```

```python
plt.figure(figsize=(8,8))
plt.plot(c[0],c[1],'o')
for i in range(10):
    c=square(9-i, (0,0))
    plt.plot(*c)
plt.show();
```

```python
plt.figure(figsize=(8,8))
plt.plot(c[0],c[1],'o')

for i in range(10):
    c=square(9-i, (0,i))
    plt.plot(*c)
plt.show();
```

```python
def square_rot(L, center=(0,0), ro=0):
    c=square(L, center)
    
    rot=np.array([
        [np.cos(ro), -np.sin(ro)],
        [np.sin(ro), np.cos(ro)]
    ])
    
    res=[]
    for i in range(len(c[0, :])):
        q=np.dot(rot, c[:, i])
        res.append(q)
        
    return np.array(res).T 
```

```python
with plt.xkcd():               # estilo xkcd
    plt.figure(figsize=(8,8))  # establecer figura y tamaÃ±o
    
    index_ticks=list(range(-20, 20, 4))    # indices eje x
    label_ticks=list(map(lambda x: 'G{}'.format(x), index_ticks)) # etiquetas eje x
    print (index_ticks, label_ticks)
    plt.xticks(index_ticks, label_ticks, rotation=55)  # cambio de xticks con rotacion
    
    for i in range(50): # 50 cuadrados
        c=square_rot(10-i, (0,0), (np.pi/50)*i)
        plt.plot(*c, color='#336699', alpha=1-(i/50))
    
plt.title('Cuadrados')  # titulo
plt.xlabel('Coord X')  # titulos de coordenadas
plt.ylabel('Coord Y')
plt.legend(['Hola que pashaaa'], loc='best') # leyenda, best es donde sea sin tapar grafico
plt.show();
```
