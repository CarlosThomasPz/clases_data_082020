```python
from pandas.io.json import json_normalize
#json_normalize desglosara los diccionarios o listas de diccionarios que esten las filas del un dataframe
#convirtiendo los diccionarios en mas columnas que se aadiran del lado dercho siguiendo con el mismo numero de filas
```

```python
lst_repo = [i['repo'] for i in resp.json()] #De igual manera se puede obtener infomracion directa del json
                                            # sin tener que usar json_normalize
```

```python
pd.DataFrame(lst_repo) 
```

```python
def get_repo(json): #Funcion para desglosar json por su repo
    lst = []
    for i in json:
        lst.append(i['repo'])
        
    df = pd.DataFrame(lst)
    return df
```

```python
repos = get_repo(resp.json())
```

# API Rick and Morty 

```python
r_m_api = req.get('https://rickandmortyapi.com/api/')
r_m_api.json() #La API nos regresa a su vez API's para los personajes, locaciones, y episodios
```

```python
r_m_char = r_m_api.json()['characters']# API para personales (url)
r_m_char
```

```python
r_m_char_com = req.get(r_m_char)#Solicitamos los personajes
```

```python
r_m_json = r_m_char_com.json()
```

```python
r_m_json.keys()# La respuesta nos regresa dos campos del cual queremos results
```

```python
df_rm = json_normalize(r_m_json['results'])
```

```python
df_rm
```

```python
df_rm.loc[1,['image']].values[0] #Como se pueden obtener las url de los personajes
```
